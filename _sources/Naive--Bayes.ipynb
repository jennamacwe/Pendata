# NAIVE BAYES

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/datasets/breast-cancer/master/data/breast-cancer.csv')
dataset.head(5)
```

```python
dataset.drop('breast-quad', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
dataset['age']=enc.fit_transform(dataset['age'].values)
dataset['mefalsepause']=enc.fit_transform(dataset['mefalsepause'].values)
dataset['tumor-size']=enc.fit_transform(dataset['tumor-size'].values)
dataset['inv-falsedes']=enc.fit_transform(dataset['inv-falsedes'].values)
dataset['falsede-caps']=enc.fit_transform(dataset['falsede-caps'].values)
dataset['deg-malig']=enc.fit_transform(dataset['deg-malig'].values)
dataset['breast']=enc.fit_transform(dataset['breast'].values)
dataset['irradiat']=enc.fit_transform(dataset['irradiat'].values)
dataset['class']=enc.fit_transform(dataset['class'].values)
dataset.head(10)
```

```python
dataset.info()
```

```python
x=dataset.drop(["class"],axis=1)
x.head()
```

```python
x_test = dataset.drop(["class"],axis=1)
x_test.head()
```

```python
y=dataset["class"]
y.head()
```

```python
from sklearn.naive_bayes import GaussianNB
modelnb = GaussianNB()
nbtrain =modelnb.fit(x,y)
Y_predict = nbtrain.predict(x_test)
print("prediksi :", Y_predict)
```

# KKN

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from seaborn import load_dataset
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasets/breast-cancer/master/data/breast-cancer.csv")
df.head()
```

```python
df.shape
```

```python
# X tanpa label
df = df.dropna()

X = df[['deg-malig']]
y = df['class']
df.head()
```

```python
#y labelnya saja
y = df["class"].values
y[0:272]
```

```python
# Splitting data into training and testing data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 100)
```

```python
KNeighborsClassifier(
    n_neighbors=5,          # The number of neighbours to consider
    weights='uniform',      # How to weight distances
    algorithm='auto',       # Algorithm to compute the neighbours
    leaf_size=30,           # The leaf size to speed up searches
    p=2,                    # The power parameter for the Minkowski metric
    metric='minkowski',     # The type of distance to use
    metric_params=None,     # Keyword arguments for the metric function
    n_jobs=None             # How many parallel jobs to run
)
```

```python
# Creating a classifier object in sklearn
clf = KNeighborsClassifier(p=1)
```

```python
# Fitting our model
clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test)
print(predictions)
```

```python
predictions = clf.predict([[44.2]])
print(predictions)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[:5]
```

```python
knn = KNeighborsClassifier(n_neighbors = 5,metric = 'minkowski', p = 2)
knn.fit(X_train,y_train)
```

```python
knn.predict(X_test)[0:5]
```

```python
knn.score(X_test, y_test)
```

```python
from sklearn.metrics import *
import numpy as np

#Check Accuracy precision, recall, f1-score
print( classification_report(y_test, knn.predict(X_test)) )
#Another way to get the models accuracy on the test data
print(F'Accuracy:',accuracy_score(y_test, knn.predict(X_test)))
```
